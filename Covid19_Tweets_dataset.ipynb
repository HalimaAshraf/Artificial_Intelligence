```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('/content/covid19_tweets.csv')
dataset
```

```python
label = LabelEncoder()
label.fit(dataset.user_location.drop_duplicates())
dataset.user_location = label.transform(dataset.user_location)

label.fit(dataset.user_description.drop_duplicates())
dataset.user_description = label.transform(dataset.user_description)

label.fit(dataset.user_verified.drop_duplicates())
dataset.user_verified = label.transform(dataset.user_verified)

label.fit(dataset.text.drop_duplicates())
dataset.text = label.transform(dataset.text)

label.fit(dataset.user_created.drop_duplicates())
dataset.user_created = label.transform(dataset.user_created)

label.fit(dataset.date.drop_duplicates())
dataset.date = label.transform(dataset.date)

label.fit(dataset.hashtags.drop_duplicates())
dataset.hashtags = label.transform(dataset.hashtags)

label.fit(dataset.source.drop_duplicates())
dataset.source = label.transform(dataset.source)

label.fit(dataset.is_retweet.drop_duplicates())
dataset.is_retweet = label.transform(dataset.is_retweet)

dataset
```

```python
from numpy.random.mtrand import RandomState
X_lin = dataset.drop(['user_favourites'], axis = 1)
Y_lin = dataset[['user_favourites']]


X_lin_train, X_lin_test, Y_lin_train, Y_lin_test = train_test_split(X_lin, Y_lin, test_size=0.3, random_state=42)
```

```python
Linear_model = LinearRegression()
Linear_model.fit(X_lin_train, Y_lin_train)
```

```python
pred = Linear_model.predict(X_lin_test)
pred
```

```python
for idx, col_name in enumerate(X_lin_train.columns):
  print("The cofficient for {} is {}".format(col_name, Linear_model.coef_[0][idx]))
```
